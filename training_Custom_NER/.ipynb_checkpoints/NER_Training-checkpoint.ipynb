```python
import spacy
import json

with open('Annotated_with_(education,phone,place).json','r') as f:
      data = json.load(f)
```

```python
data[0]['label']
```

```python
training_data = []

for example in data:
    temp_dic = {}
    temp_dic['text'] = example['text']
    temp_dic['entities'] = []
    for annotation in example['label']:
        start = annotation[0]
        end = annotation[1]
        label = annotation[2]
        temp_dic['entities'].append((start,end,label))
    training_data.append(temp_dic)
```

```python
training_data
```

```python
len(training_data)
```

```python
from spacy.tokens import DocBin
from tqdm import tqdm
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(training_data, test_size=0.2)
```

```python
nlp = spacy.blank('en')
doc_bin = DocBin()
```

```python
from spacy.util import filter_spans

def get_spacy_doc(data):
    training_data = data
    for training_example in tqdm(training_data):
        text = training_example['text']
        labels = training_example['entities']
        doc = nlp.make_doc(text)
        ents = []
        # print(doc)

        for start,end,label in labels:
            span = doc.char_span(start,end,label, alignment_mode='contract')
            if span is None:
                print('Skipping entity')
            else:    
                # print(span)
                ents.append(span)

        filtered_ents = filter_spans(ents)
        doc.ents = filtered_ents
        doc_bin.add(doc)
    return doc_bin
```

```python
doc_bin = get_spacy_doc(train)
doc_bin.to_disk('train.spacy')


doc_bin = get_spacy_doc(test)
doc_bin.to_disk('test.spacy')
```
